```python
import torch
import torchvision
import torchvision.utils as vutils
import torchvision.datasets as datasets
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
```

```python
def show_image(dataset, batch_size):
    transformation = transforms.Compose([transforms.ToTensor()])
    if dataset == 'cifar10':
        trainset = datasets.CIFAR10(root='./data', train=True, transform=transformation, download=True)
    elif dataset == 'cifar100':
        trainset = datasets.CIFAR10(root='./data', train=True, transform=transformation, download=True)
    elif dataset == 'svhn':
        trainset = datasets.SVHN(root='./data', split='train', transform=transformation, download=True)
        
    loader = torch.utils.data.DataLoader(trainset, batch_size=batch_size, shuffle=False)
    images = next(iter(loader))[0]
    grid = vutils.make_grid(images, padding=2)
    
    plt.figure(figsize=(10, 10))
    plt.imshow(grid.permute(1, 2, 0))
```

```python
show_image('svhn', 64)
```

```python
show_image('cifar10', 64)
```
